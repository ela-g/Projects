```python
# importing 'random' module to assign turns randomly in each round. 
import random

# importing 'getpass' to hide the weapon inputs on the screen (from the other player)
import getpass

# a func that receives to weapons and returns the scores of the the weapons' holders
# in one Rock-Paper-Scissors battle.
def rps (weapon_1, weapon_2):
    score_1 = 0
    score_2 = 0
    # in this 3 combinations the holder of the first weapon wins
    first_wins_combinations = [('Rock', 'Scissors'), ('Scissors', 'Paper'), ('Paper', 'Rock')]
    if (weapon_1, weapon_2) in first_wins_combinations: 
        score_1 = 1
    
    # the lambda reverses one combination and the map appleies it to all 3 combinations - 
    # now they refer to a victory of second weapon
    elif (weapon_1, weapon_2) in list(map(lambda x: tuple(reversed(x)), first_wins_combinations)):
        score_2 = 1
    # each score refers to the holder of the weapon which has the same index in the func input tuple  
    return(score_1, score_2)

player_a_name = input("Hi! The player on the right - yes, YOU - what is your name?\t").capitalize()
player_b_name = input("And you on the left, what\'s yours?\t").capitalize()

# Validating that the names are different (because they will be dict.keys). Not case sensitive, to avoid confusion.
while player_b_name.lower() == player_a_name.lower():
    player_b_name = input('Sorry, {} is already taken. \nMaybe you have a second name or a nickname?\t'\
          .format(player_b_name.capitalize()))
    
winning_score = input("up to which score do you want to play?\t")

# creating a dictionary with the game's data. 
# The main keys of the dictionary will be the indexes of the rounds.
# Each value will be a dictionary by itself with the 4 items: 
# key - first player by name; value - a list of weapo—Ç, current round score and total score
# key - second player by name; value - same 
# key - 'round winner'; value - name
# key - 'round loser'; value - name 
game_details = {}

# playing until reaching the 'winning_score' score. Each iteration is a round.
while not winning_score.isnumeric():
    winning_score = input("This wasn\'t a number. Try again:\nup to which score do you want to play?\t" )    
    
i = 0
total_player_a_score = 0
total_player_b_score = 0
leading_score = 0


while leading_score < int(winning_score):
    
    # assigning the dict keys
    game_details[i] = {player_a_name: [], player_b_name: []}
    
    # chosing randomly which plyer will chose a weapon first in each round, to make the game less repetitive...
    firts_in_this_round = random.choice(list(game_details[i].keys()))
    second_in_this_round = [key for key in game_details[i].keys() if key != firts_in_this_round][0]
    
    # chosing weapons by players. Full word or just the the fitst lette, upper or lower case - are all good.
    legit_weapons_lower = ['rock', 'r', 'paper', 'p', 'scissors', 's']
    
    # the players input(weapon) will appear on the screen as a black dot. 
    firts_in_this_round_weapon = getpass.getpass(prompt = "\n\033[;1m{}\033[0;0m, what is your weapon? \nYou can use Rock(r), Paper(p) or Scissors(s)"\
                                                         .format(firts_in_this_round))
    # validating that the 'weapon' input is ok, and asking to chose again if not
    while firts_in_this_round_weapon.lower() not in legit_weapons_lower:
        firts_in_this_round_weapon = getpass.getpass(prompt = 
        "\nOops. This wasn\'t a legit weapon. \nYou can only use \033[;1m Rock(r), Paper(p) or Scissors(s) \033[0;0m. \n\nwhat is your weapon?")
    
    second_in_this_round_weapon = getpass.getpass(prompt = "\n\033[;1m{}\033[0;0m, what is your weapon? \nYou can use Rock(r), Paper(p) or Scissors(s)"\
                                                          .format(second_in_this_round))
    
    while second_in_this_round_weapon.lower() not in legit_weapons_lower:
        second_in_this_round_weapon = getpass.getpass(prompt = 
        "\nOops. This wasn\'t a legit weapon. \nYou can only use \033[;1m Rock(r), Paper(p) or Scissors(s) \033[0;0m. \n\nwhat is your weapon?")
    
    # a small func to convert legit weapons to have one uniformed form - so it would look good when printing  
    uni_weapon_name = lambda x: 'Rock' if x.lower() in ['r', 'Rock'] else\
                    ('Paper' if x.lower() in ['p', 'paper'] else 'Scissors') 
    
    # assigning the weapons to their owners - adding the first element in the values of the items that
    # go by the key of players names.
    game_details[i][firts_in_this_round].append(uni_weapon_name(firts_in_this_round_weapon))
    game_details[i][second_in_this_round].append(uni_weapon_name(second_in_this_round_weapon))
    
    # assigning thw weapons that were already added to the dict, to variables outside the dict - 
    # just to make it easier to use the rps function to get the scores
    (player_a_weapon, player_b_weapon) = (game_details[i][player_a_name][0], game_details[i][player_b_name][0])
    
    (player_a_score, player_b_score) = rps(player_a_weapon, player_b_weapon)
    
    # adding the func output as the second element in the value of each player's dict
    # (of the two dicts that have a key of player's name)
    game_details[i][player_a_name].append(player_a_score)
    game_details[i][player_b_name].append(player_b_score)
    
    total_player_a_score += player_a_score
    total_player_b_score += player_b_score
    
    # adding the total score as the third element in the value of each player's dict
    # (of the two dicts that have a key of player's name)
    game_details[i][player_a_name].append(total_player_a_score)
    game_details[i][player_b_name].append(total_player_b_score)
    
    # calculating who wins/loses the current round and who is leading/losing the game by now.
    leading_score = max(total_player_a_score, total_player_b_score)
    
    round_winner = [player for player, [weapon, round_score, total_score] in game_details[i].items()\
                  if round_score == 1]

    round_loser = [player for player, [weapon, round_score, total_score] in game_details[i].items()\
                  if round_score == 0 and round_winner]

    mid_winner = [player for player, [weapon, round_score, total_score] in game_details[i].items()\
                  if total_score == leading_score and total_player_a_score != total_player_b_score]

    mid_loser = [player for player, [weapon, round_score, total_score] in game_details[i].items()\
                  if total_score != leading_score and total_player_a_score != total_player_b_score]
    
    # extracting the relevant name from the list if there is a name, and leaving it as an empty
    # list in case there is no winner
    extract = lambda x: x[0] if x else x
    (round_winner, round_loser, mid_winner, mid_loser) = map(extract, [round_winner, round_loser, mid_winner, mid_loser])
    
    # assigning the values to the dicts of round winner and round loser. This dicts' value is a single string
    game_details[i]['round winner'] = round_winner if round_winner else 'No one'
    game_details[i]['round loser'] = round_loser if round_loser else 'No one'
    
    # printing the first part of the round summary that tells the result of the current round-
    # in case ther is and in case there isn't a round winner
    if not round_winner:
        print('\n\033[4mROUND {}:\033[0m You had the same weapon,\n\t\033[;1m {} wins this round\033[0m'\
              .format(i+1, game_details[i]['round winner']))
    else:
        print('\n\033[4mROUND {}:\033[0m \033[;1m {} wins this round\033[0m with {} over {}'\
              .format(i+1, round_winner, game_details[i][round_winner][0], game_details[i][round_loser][0]))
        
    # printing the score summary by now - in case someone leads and in case both players are even
    if not mid_winner:
        print('{:>16} \033[1;34m you are EVEN with {} point{} \033[0m'\
              .format('By now', leading_score, '' if leading_score == 1 else 's'))
    elif leading_score == int(winning_score):
        final_winner = mid_winner
        print('\n{:>9}\033[;1m \033[1;36m{} IS THE WINNER!!!\033[0;0m'.format(' ',final_winner.upper()))
    else:
        print('{:>16} \033[1;36m{} {} point{}\033[0;0m \n{:>17}\033[1;31m{} {} point{}\033[0;0m'\
              .format('By now', mid_winner, leading_score, '' if leading_score == 1 else 's',\
                      ' ', mid_loser, game_details[i][mid_loser][2], '' if game_details[i][mid_loser][2] == 1 else 's'))
    
    # getting ready for the next iteration
    i += 1


to_see_resume = input('\nPress Y to see the game summary\t')

if to_see_resume.lower() == 'y':
    print('\nAll in all there were {} rounds'.format(len(game_details.keys())))
    print('{} won {} rounds and {} won {} rounds.'.format(player_a_name, total_player_a_score,\
                                                           player_b_name, total_player_b_score))
    for i in game_details.keys():
        print('Round {:>2} : {:^8} won  -  {:<6} had {:<8} and {:<6} had {:<8}'\
              .format(i+1, round_winner, round_winner, game_details[i][round_winner][0],\
                      round_loser, game_details[i][round_loser][0]))
```
